```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
import scipy.stats as stats
import seaborn as sns
```

```python
base = pd.read_csv('3.Algoritmos de Machine Learning\mt_cars.csv')
base.shape
```

```python
base.head()
```

```python
base = base.drop('Unnamed: 0', axis = 1)
```

```python
base.head()
```

```python
corr = base.corr()
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt='.2f')
```

```python
col = [('mpg', 'cyl'), ('mpg', 'disp'), ('mpg', 'hp'), ('mpg', 'wt'), ('mpg', 'drat'), ('mpg', 'vs')]
n_plots = len(col)
fig, axes = plt.subplots(nrows=n_plots, ncols=1, figsize=(6, 4 * n_plots))

for object, pair in enumerate(col):
    x_col, y_col = pair
    sns.scatterplot(x=x_col, y=y_col, data = base, ax=axes[object])
    axes[object].set_title(f'{x_col} vs {y_col}')

plt.tight_layout()
plt.show()
```

```python
#aic 156,6, bic 162.5
modelo = sm.ols(formula='mpg ~ wt + disp + hp', data=base)
modelo = modelo.fit()
modelo.summary()
```

```python
residuos = modelo.resid
plt.hist(residuos, bins=20)
plt.xlabel("Residuos")
plt.ylabel("Frequencia")
plt.title("Histogramas de residuos")
```

```python
stats.probplot(residuos, dist="norm", plot=plt)
plt.title("Q-Q Plot de Residuos")
plt.show()
```

```python
# h0 - dados esto normalmente distribudos
# p < 0.05 rejeito a hiptese nula, (no esto distribuidos)
# p > 0.05 no  possivel rejeitar o h0

stat, pval = stats.shapiro(residuos)
print(f'Shapiro-wilk: {stat:.3f}, p-value: {pval:.3f}')
```
